# Importing Library & Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# Load and Prepare Data

```python
df = pd.read_csv("HP_DRUM_LEVEL_revise-Mean-Std-MSE-Range.csv")
df
```

# Create Condition

```python
df_label1 = df[df["Label"] == 1]
df_label1
```

```python
global df_mean, start_range , end_range
df_mean = []
start = -700
end = 900
start_range = []
end_range = []

while start <= 800:
    df_mean.append(df_label1[(df_label1["Mean"]>=start) & (df_label1["Mean"]<start+100)])
    start_range.append(start)
    end_range.append(start+100)
    start = start + 100
```

```python
for i in range(len(df_mean)):
    print("{}_{}: {}".format(start_range[i], end_range[i], df_mean[i]["Standard Deviation"].min()))
```

```python
for i in range(len(df_mean)):
    print("{}_{}: {}-{}".format(start_range[i], end_range[i], df_mean[i]["MSE"].min(),
                               df_mean[i]["MSE"].max()))
```

```python
def classification(row):
    if row['Mean'] < -400 and row['Standard Deviation'] >= 18.46259137:
        if (row['Mean'] >= -600 and row['Mean'] < -500 and row['MSE'] >= 272.76248 and
           row['Standard Deviation'] <= 371.823216):
            if row['Range'] >= 50.02:
                return 1
            else: return 0
        elif (row['Mean'] >= -500 and row['Mean'] < -400 and row['MSE'] >= 273.027416 and
           row['MSE'] <= 410.215704):
            if row['Range'] >= 50.02:
                return 1
            else: return 0
        return 1
    elif row['Mean'] >= -400 and row['Mean'] < -300 and row['Standard Deviation'] >= 18.57627035:
        if (row['MSE'] >= 270.608504 and row['MSE'] <= 463.478504):
            if row['Range'] >= 50.02:
                return 1
            else: return 0
        return 1
    elif row['Mean'] >= -300 and row['Mean'] < -200 and row['Standard Deviation'] >= 18.39063811:
        if (row['MSE'] >= 276.069624 and row['MSE'] <= 487.078576):
            if row['Range'] >= 50.02:
                return 1
            else: return 0
        elif (row['MSE'] >= 270.608504 and row['MSE'] <= 276.059416):
            return 0
        return 1
    elif row['Mean'] >= -200 and row['Mean'] < -100 and row['Standard Deviation'] >= 17.77051687:
        if (row['MSE'] >= 252.637064 and row['MSE'] <= 406.84484):
            if row['Range'] >= 50.02:
                return 1
            else: return 0
        return 1
    elif row['Mean'] >= -100 and row['Mean'] < 0 and row['Standard Deviation'] >= 18.21877877:
        if (row['MSE'] >= 265.570536 and row['MSE'] <= 429.38772):
            if row['Range'] >= 50.02:
                return 1
            else: return 0
        return 1
    elif row['Mean'] >= 0 and row['Mean'] < 100 and row['Standard Deviation'] >= 18.33792655:
        if (row['MSE'] >= 269.03228 and row['MSE'] <= 391.654744):
            if row['Range'] >= 50.02:
                return 1
            else: return 0
        return 1
    elif row['Mean'] >= 100 and row['Mean'] < 200 and row['Standard Deviation'] >= 19.44528786:
        if (row['MSE'] >= 302.51252 and row['MSE'] <= 395.447784):
            if row['Range'] >= 50.02:
                return 1
            else: return 0
        return 1
    elif row['Mean'] >= 200 and row['Mean'] < 300 and row['Standard Deviation'] >= 18.54825275:
        if (row['MSE'] >= 275.234344 and row['MSE'] <= 460.250816):
            if row['Range'] >= 50.02:
                return 1
            else: return 0
        return 1
    elif row['Mean'] >= 300 and row['Mean'] < 400 and row['Standard Deviation'] >= 19.31149839:
        if (row['MSE'] >= 298.351416 and row['MSE'] <= 380.653336):
            if row['Range'] >= 50.02:
                return 1
            else: return 0
        return 1
    elif row['Mean'] >= 400 and row['Standard Deviation'] >= 18.44909293:
        if (row['Mean'] >= 400 and row['Mean'] < 500 and row['MSE'] >= 272.301376 and
           row['Standard Deviation'] <= 346.681616):
            if row['Range'] >= 50.02:
                return 1
            else: return 0
        elif (row['Mean'] >= 500 and row['Mean'] < 600 and row['MSE'] >= 272.48076 and
           row['MSE'] <= 356.87104):
            if row['Range'] >= 50.02:
                return 1
            else: return 0
        elif (row['Mean'] >= 600 and row['Mean'] < 700 and row['MSE'] >= 272.360744 and
           row['MSE'] <= 382.9914):
            if row['Range'] >= 50.02:
                return 1
            else: return 0
        elif (row['Mean'] >= 700 and row['Mean'] < 800 and row['MSE'] >= 272.786056 and
           row['MSE'] <= 386.853816):
            if row['Range'] >= 50.02:
                return 1
            else: return 0
        elif (row['Mean'] >= 800 and row['Mean'] < 900 and row['MSE'] >= 272.637216 and
           row['MSE'] <= 321.971416):
            if row['Range'] >= 50.02:
                return 1
            else: return 0
        return 1
    else: return 0
```

```python
df['Classification'] = df.apply(lambda row: classification(row), axis=1)
df
```

```python
from sklearn.metrics import mean_absolute_error

test_mae = mean_absolute_error(df['Label'], df['Classification'])
# print(test_mae)
print("Mean Absolute Error: {}".format(test_mae))
# print("Mean test data value: {}".format(np.mean(df['Classification'])))
```

```python
confusion_matrix = pd.crosstab(df['Label'], df['Classification'], rownames=['Actual'], colnames=['Predicted'])
print(confusion_matrix)
```

```python
from sklearn.metrics import accuracy_score

acc = accuracy_score(df['Label'], df['Classification'])
print("Accuracy score: {}".format(acc))
# acc
```
